```python
# Loading libraries

import pandas as pd
import numpy as np
```

```python
# -------------------------------
# Step 1: Load Data
# -------------------------------
df = pd.read_csv('sales_data.csv')

print("Initial Data:")
print(df)
```

```python
# -------------------------------
# Step 2: Explore Data
# -------------------------------
print("\nShape of dataset:", df.shape)
print("\nColumns:", df.columns)
print("\nData Types:\n", df.dtypes)
```

# Handling missing values

```python
df.isnull().sum()
```

# Handling duplicate values

```python
df.duplicated().sum()
```

# Analyzing data

```python
# Total Revenue

total_revenue = sum(df['Total_Sales'])
```

```python
total_revenue
```

# Checking best product sold based on multiple categories

```python
# Based on quantity

best_product_quantity = df.groupby('Product')['Quantity'].idxmax()
```

```python
best_product_quantity
```

```python
# Based on price

best_product_price = df.groupby('Product')['Price'].idxmax()
```

```python
best_product_price
```

```python
# Based on revenue generated

top_revenue_product = df.groupby('Product')['Total_Sales'].idxmax()
```

```python
top_revenue_product
```
