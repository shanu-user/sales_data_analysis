```python
# Loading libraries

import pandas as pd
import numpy as np
```

```python
# -------------------------------
# Step 1: Load Data
# -------------------------------
df = pd.read_csv('sales_data.csv')

print("Initial Data:")
print(df)
```

```python
# -------------------------------
# Step 2: Explore Data
# -------------------------------
print("\nShape of dataset:", df.shape)
print("\nColumns:", df.columns)
print("\nData Types:\n", df.dtypes)
```

# Analyzing data

```python
# Total Revenue

total_revenue = sum(df['Total_Sales'])
```

```python
total_revenue
```

# Checking best product sold based on multiple categories

```python
# Based on quantity

best_product_quantity = df.groupby('Product')['Quantity'].sum().sort_values(ascending=False)
```

```python
best_product_quantity
```

```python
# Based on sales generated

top_sales_product = df.groupby('Product')['Total_Sales'].sum().sort_values(ascending=False)
```

```python
top_sales_product
```

```python
product_revenue = df.groupby('Product')['Total_Sales'].sum()
```

```python
product_revenue
```

```python
total_revenue = product_revenue.sum()
```

```python
total_revenue
```

```python
revenue_percentage = (product_revenue / total_revenue) * 100
```

```python
revenue_percentage
```

```python
revenue_summary = pd.DataFrame({
    'Total_Revenue': product_revenue,
    'Contribution_%': revenue_percentage
}).sort_values(by='Total_Revenue', ascending=False)

print(revenue_summary)
```
